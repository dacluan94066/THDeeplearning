```python
# cau8:
rd_array = np.random.randint(0, 21, size=20)
print("Mảng gốc:", rd_array)


condition = (rd_array >= 9) & (rd_array <= 15)

modified_array = rd_array.copy()
modified_array[condition] = -modified_array[condition]

print("Mảng sau khi đổi dấu:", modified_array)
```
